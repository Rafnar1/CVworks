```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from utils.image_processing import resize_image_keep_ratio
from utils.features.sift import generate_dogs, convolve_with_dogs, non_maximum_supression
```

```python
MAX_HEIGHT = 300
MAX_WIDTH = 300

INITIAL_SIGMA = 3
INITIAL_K = 1.2
INITIAL_KERNEL_SIZE = 11
```

```python
# TODO: Implemenet the generate_dogs function

dogs = generate_dogs(6, INITIAL_KERNEL_SIZE, INITIAL_K, INITIAL_SIGMA)
for dog in dogs:
    print(dog.shape)
    plt.figure(figsize=(4, 4))
    plt.imshow(dog, cmap='gray')
    plt.show()
```

```python
image = cv2.cvtColor(cv2.imread('image.jpeg'), cv2.COLOR_BGR2GRAY)
image = resize_image_keep_ratio(image, MAX_HEIGHT, MAX_WIDTH)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray')
```

```python
# TODO: Implemenet the convolve_with_dogs function and analyze blobs in the outputs

convolved_images = convolve_with_dogs(image, dogs)
for convolved_image in convolved_images:
    plt.imshow(convolved_image, cmap='gray')
    plt.show()
```

```python
# TODO: Implemenet the non_maximum_supression function to filter noises

convolved_images_filtered = non_maximum_supression(np.array(convolved_images), -3)
for convolved_image_filtered in convolved_images_filtered:
    
    plt.imshow(convolved_image_filtered, cmap='gray')
    plt.show()
```
