```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

from utils.image_processing import resize_image_keep_ratio
from utils.clustering.kmeans import kmeans
```

```python
MAX_HEIGHT = 500
MAX_WIDTH = 500
```

```python
image = cv2.cvtColor(cv2.imread('nature.jpeg'), cv2.COLOR_BGR2RGB)
image = resize_image_keep_ratio(image, MAX_HEIGHT, MAX_WIDTH)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```

```python
# TODO: Implemenet the kmeans function
centers, clustered = kmeans(image, 5)

plt.figure(figsize=(10, 10))
plt.imshow(clustered, cmap='gray')
plt.show()
```

```python
# TODO: Implemenet the kmeans function
centers, clustered = kmeans(image, 5, include_position=True)

plt.figure(figsize=(10, 10))
plt.imshow(clustered, cmap='gray')
plt.show()
```
