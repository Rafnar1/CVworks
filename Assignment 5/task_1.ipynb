```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
```

```python
np.random.seed(0)
```

```python
def display_sample(sample):
    plt.figure(figsize=(2, 2))
    plt.imshow(sample, cmap='gray')
    plt.show()
```

```python
tf.keras.datasets.mnist.load_data(path="mnist.npz")

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
assert X_train.shape == (60000, 28, 28)
assert X_test.shape == (10000, 28, 28)
assert y_train.shape == (60000,)
assert y_test.shape == (10000,)
```

```python
print("Train X data shape: %s" % (X_train.shape,))
print("Train y data shape: %s" % (X_train.shape,))
print("Test X data shape: %s" % (X_test.shape,))
print("Test y data shape: %s" % (y_test.shape,))
```

```python
for i in range(10):
    random_idx = np.random.randint(0, len(X_test))
    display_sample(X_train[random_idx])
    print("Label: %s" % y_train[random_idx])
```

```python
X_train_flatten = np.reshape(X_train, (X_train.shape[0], X_train.shape[1] * X_train.shape[2]))
X_test_flatten = np.reshape(X_test, (X_test.shape[0], X_test.shape[1] * X_test.shape[2]))
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

# TODO: Use Logistic Regression with l1 penalty for the classification of digits

# Create a Logistic Regression model
model = LogisticRegression(solver='liblinear', max_iter=1000, penalty='l1')

# Train the model
model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model with L2 penalty
model = LogisticRegression(penalty='l2', max_iter=1000)

# Train the model
model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```

```python
from sklearn.svm import LinearSVC

# TODO: Use Support Vector Machines with l2 penalty for the classification of digits

# Create a LinearSVC model with L2 penalty
model = LinearSVC(penalty='l2', dual=False, max_iter=1000)

# Train the model
model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# TODO: Use K Neares Neighbors Classifier (k = 10) for the classification of digits

model = KNeighborsClassifier(n_neighbors=10)

# Train the model
model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

# TODO: Use Decision Tree Classifier for the classification of digits

model = DecisionTreeClassifier()

model.fit(X_train_flatten, y_train)

print("Accuracy (train): %s" % model.score(X_train_flatten, y_train))
print("Accuracy (test): %s" % model.score(X_test_flatten, y_test))
```
